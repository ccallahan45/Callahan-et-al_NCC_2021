Set boxplot color function

```python
def set_boxplot_col(b,n,c,c_double,lw,median):
    
    # This function sets the colors of a given boxplot b
    # It auto-sets the colors of the boxes, whiskers, caps, and fliers to c[m]
    # and it sets the colors of the medians if median=True
    # n is the number of instances of the boxplot and lw is the desired
    # linewidth
    # c should be a vector of colors
    
    for m in range(0,n*2,1):
        if m < n:
            b["boxes"][m].set(color=c[m],linewidth=lw)
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
            #b["fliers"][m].set(markerfacecolor=c[m],markeredgecolor=c[m])
            if median==True:
                b["medians"][m].set(color=c[m],linewidth=lw)
        else:
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
```

#### Analysis/plotting

```python
transient_change_nino3 = np.zeros(len(modelnames_file))
equilibrium_change_nino3 = []
#pvals = np.zeros(len(modelnames_file))

transient_change_nino4 = np.zeros(len(modelnames_file))
equilibrium_change_nino4 = []
```

```python
rm_yrs = 100
```

```python
for i in np.arange(0,len(modelnames_file),1):
    
    if modelnames_file[i] == "MPIESM11_abrupt4x":
        rm_length = rm_yrs
    else:
        rm_length = rm_yrs*12
        
    model, exp = modelnames_file[i].split("_")
    print(modelnames_file[i])
    
    for ii in np.arange(0,2,1):
        if ii == 0:
            index_ctrl_anom = xr.DataArray(xr.open_dataset(loc_in+"nino3_"+model+"_control_anom_detrend2.nc").data_vars["nino3"])
            index_forced_anom = xr.DataArray(xr.open_dataset(loc_in+"nino3_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino3"])
        else:
            index_ctrl_anom = xr.DataArray(xr.open_dataset(loc_in+"nino4_"+model+"_control_anom_detrend2.nc").data_vars["nino4"])
            index_forced_anom = xr.DataArray(xr.open_dataset(loc_in+"nino4_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino4"])
    
        std_f1 = pd.Series(index_forced_anom).rolling(rm_length,center=False).std().values
        std_c1 = pd.Series(index_ctrl_anom).rolling(rm_length,center=False).std().values

        std_c = std_c1[~np.isnan(std_c1)]
        std_f = std_f1[~np.isnan(std_f1)]

        baseline1 = np.mean(std_c)
        baseline2 = np.std(index_ctrl_anom)

        amp_change = ((std_f - baseline1)/baseline1)*100
        amp_cdist = ((std_c - baseline1)/baseline1)*100
        
        if runtype[i] == "abrupt":
            amp_change_transient = np.mean(amp_change[(50*12):(150*12)])
            amp_change_eq = amp_change[(150*12):]

        else:
            amp_change_transient = np.mean(amp_change[0:(140*12)])
            amp_change_eq = amp_change[(140*12):]

        if ii == 0:
            transient_change_nino3[i] = amp_change_transient
            equilibrium_change_nino3.append(amp_change_eq)
        else:
            transient_change_nino4[i] = amp_change_transient
            equilibrium_change_nino4.append(amp_change_eq)
```
