Custom detrending function

```python
def custom_detrend(data,order):
    
    # only for one-dimensional timeseries
    # numpy required
    
    x = np.arange(1,len(data)+1,1)
    
    model = np.polyfit(x,data,order)
    predicted = np.polyval(model,x)
    new = data - predicted
    
    return(new)
```

#### Analysis

Create figure object, loop through, plot GMST/ENSO and the distributions of ENSO amplitudes below it

```python
rm_yr = 100
rm_length = rm_yr*12
```

```python
fig = plt.figure(figsize=(18,16))

rcParams["font.family"] = "sans-serif"
rcParams["font.sans-serif"] = ["Helvetica Neue"]
rcParams["font.size"] = 26
rcParams["axes.linewidth"] = 1.5
rcParams['xtick.major.size'] = 8
rcParams['xtick.major.width'] = 1.5
rcParams['ytick.major.size'] = 8
rcParams['ytick.major.width'] = 1.5

# originally top 0.9, bottom 0.47
# and originally top 0.37, bottom 0.1

gs1 = gridspec.GridSpec(1,3)
gs1.update(left=0.12,right=0.96,top=0.53,bottom=0.1,wspace=0.05,hspace=0.42)

gs2 = gridspec.GridSpec(1,3)
gs2.update(left=0.12,right=0.96,top=0.9,bottom=0.63,wspace=0.05,hspace=0.42)

for i in np.arange(0,len(ex),1):

    print(ex1[i])
    model, exp = ex1[i].split("_")
    
    # GMST
    gmst_anom = xr.DataArray(xr.open_dataset(loc_gmst+"GMST_anom_"+model+"_"+exp+".nc").data_vars["gmst"])

    # Nino3.4
    nino34_ctrl_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_control_anom_detrend2.nc").data_vars["nino34"])
    nino34_forced_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino34"])
    
    
    print("Calculating amplitude change...")
    
    std_f1 = pd.Series(nino34_forced_anom).rolling(rm_length,center=False).std().values
    std_c1 = pd.Series(nino34_ctrl_anom).rolling(rm_length,center=False).std().values
    gmst_anom1 = pd.Series(gmst_anom).rolling(rm_length,center=False).mean().values
    
    std_c = std_c1[~np.isnan(std_c1)]
    std_f = std_f1[~np.isnan(std_f1)]
    gmst_anom_2 = gmst_anom1[~np.isnan(gmst_anom1)]
    
    baseline = np.mean(std_c)
    amp_change = ((std_f - baseline)/baseline)*100
    amp_cdist = ((std_c - baseline)/baseline)*100
    
    
    # Split up amplitude change and GMST
    if runtype[i] == "abrupt":
        amp_change_eq = amp_change[150*12:]
        amp_change_trns = amp_change[(50*12):(150*12)]
        amp_change_fast = amp_change[0:(50*12)]
        
        gmst_eq = gmst_anom_2[150*12:]
        gmst_trns = gmst_anom_2[(50*12):(150*12)]
        gmst_fast = gmst_anom_2[0:(50*12)]
        
    else:
        amp_change_eq = amp_change[140*12:]
        amp_change_trns = amp_change[0:(140*12)]
        amp_change_fast = amp_change[0:1]
        
        gmst_eq = gmst_anom_2[140*12:]
        gmst_trns = gmst_anom_2[0:(140*12)]
        gmst_fast = gmst_anom_2[0:1]
    
    mean_change_eq = np.mean(amp_change_eq)
    mean_change_trns = np.mean(amp_change_trns)
    
    # Now plot ENSO / GMST
    ax = plt.subplot(gs1[0,i])
    
    plt.plot(gmst_fast,amp_change_fast,color="gray",linewidth=4,label="Initial")
    plt.plot(gmst_trns,amp_change_trns,color="red",linewidth=4,label="Transient") #label="CMIP5"+"\n"+"Analogy")
    plt.plot(gmst_eq,amp_change_eq,color="black",linewidth=4,label="Equilibrium") #label="LongRunMIP"+"\n"+"Gain")
    
    plt.axhline(y=mean_change_eq,linewidth=2,color="black",linestyle="--")
    plt.axhline(y=mean_change_trns,linewidth=2,color="red",linestyle="--")
    
    # Format and add bars near y axis
    gmst_diff = np.max(gmst_anom_2) - np.min(gmst_anom_2)
    addtl_window = gmst_diff*0.1
        
    plt.xlim([np.min(gmst_anom_2)-addtl_window,np.max(gmst_anom_2)+0.1])
    
    std_plus = np.mean(amp_cdist) + np.std(amp_cdist)
    std_minus = np.mean(amp_cdist) - np.std(amp_cdist)
    gmst_min = np.min(gmst_anom_2)
    gmst_max = np.max(gmst_anom_2)
    pctile95 = np.percentile(amp_cdist,95)
    pctile05 = np.percentile(amp_cdist,5)
    
    if i != 1:
        plt.plot([gmst_min-(addtl_window*0.4),gmst_min-(addtl_window*0.8)],[pctile95,pctile95],color="black",linewidth=2,linestyle="-")
        plt.plot([gmst_min-(addtl_window*0.4),gmst_min-(addtl_window*0.8)],[pctile05,pctile05],color="black",linewidth=2,linestyle="-")
        plt.plot([gmst_min-(addtl_window*0.6),gmst_min-(addtl_window*0.6)],[pctile05,pctile95],color="black",linewidth=2,linestyle="--")
        plt.plot([gmst_min-(addtl_window*0.6),gmst_min-(addtl_window*0.6)],[std_minus,std_plus],color="black",linewidth=4,linestyle="-")
    else:
        plt.plot([gmst_min-(addtl_window*1.2),gmst_min-(addtl_window*1.6)],[pctile95,pctile95],color="black",linewidth=2,linestyle="-")
        plt.plot([gmst_min-(addtl_window*1.2),gmst_min-(addtl_window*1.6)],[pctile05,pctile05],color="black",linewidth=2,linestyle="-")
        plt.plot([gmst_min-(addtl_window*1.4),gmst_min-(addtl_window*1.4)],[pctile05,pctile95],color="black",linewidth=2,linestyle="--")
        plt.plot([gmst_min-(addtl_window*1.4),gmst_min-(addtl_window*1.4)],[std_minus,std_plus],color="black",linewidth=4,linestyle="-")
    
    if i == 0:
        plt.text(gmst_min-(addtl_window*0.25),pctile05-0.5,"5",fontsize=16)
        plt.text(gmst_min-(addtl_window*0.25),pctile95-0.5,"95",fontsize=16)
        plt.text(gmst_min-(addtl_window*0.45),std_plus-0.5,"1SD",fontsize=16)
        plt.text(gmst_min-(addtl_window*0.45),std_minus-0.5,"-1SD",fontsize=16)
        
        
    # Other formatting
    ylimits = [-55,25]
    plt.ylim(ylimits)
    
    if i == 0:
        plt.ylabel("$\Delta$ Amplitude (%)")
    else:
        plt.yticks([])
    
    if i == 1:
        plt.xlabel("Global mean temperature anomaly ($\degree$C)")
        plt.xlim([4,6.2])
        plt.xticks([4.5,5.0,5.5,6.0])
        
    #plt.title(ex[i])
    plt.axhline(y=0,linewidth=2,color="k")
    
    # Red and gray shading
    
    ax.fill_between([np.min(gmst_trns),np.max(gmst_trns)],np.min(ylimits),np.max(ylimits),color="red",alpha=0.1)
    ax.fill_between([np.min(gmst_eq),np.max(gmst_eq)+0.4],np.min(ylimits),np.max(ylimits),color="gray",alpha=0.1)
    
    # Text at top
    if i == 0:
        plt.text(4.83,21.7,"Transient",fontsize=16)
        plt.text(5.7,21.7,"Equilibrium",fontsize=16)
    #elif i == 1:
        #plt.text(2.92,21.5,"Transient",fontsize=16)
        #plt.text(3.55,21.5,"Equilibrium",fontsize=16)
    #else:
        #plt.text(5,21.5,"Transient",fontsize=16)
        #plt.text(6,21.5,"Equilibrium",fontsize=16)
    
    # Legend
    if i == 0:
        plt.legend(frameon=False,loc="center right",fontsize=16,bbox_to_anchor=(1.0,0.79))
    
    if i == 2:
        ax2 = ax.twinx()
        plt.plot(gmst_trns,amp_change_trns,color="white",linewidth=0)
        plt.ylabel("")
        plt.ylim(ylimits)
        plt.yticks(ticks=[-50,-40,-30,-20,-10,0,10,20],labels=[])
    
    
    # Distributions!
    
    ax1 = plt.subplot(gs2[0,i])
    
    if runtype[i] == "abrupt":
        amp_eq = std_f[150*12:]
        amp_trns = std_f[(50*12):(150*12)]
        amp_fast = std_f[0:(50*12)]
                
    else:
        amp_eq = std_f[140*12:]
        amp_trns = std_f[0:(140*12)]
        amp_fast = std_f[0:1]    
    
    if i == 0:
        d1 = sns.distplot(list(std_c),color="black",hist=False,kde=True,kde_kws={"linewidth":3,"alpha":1,"linestyle":"--"},label="Control")
        #p2 = plt.axvline(x=np.mean(amp_fast),color="gray",linewidth=3,linestyle="--",label="Fast (Pre-Transient)")
        p1 = plt.axvline(x=np.mean(amp_trns),color="red",linewidth=3,linestyle="-",label="Transient")
        d2 = sns.distplot(list(amp_eq),color="black",hist=False,kde=True,kde_kws={"linewidth":3,"alpha":1},label="Equilibrium")
        leg = plt.legend(frameon=False,loc="upper right",fontsize=16)
    else:
        d1 = sns.distplot(list(std_c),color="black",hist=False,kde=True,kde_kws={"linewidth":3,"alpha":1,"linestyle":"--"})#,label="Control")
        #p2 = plt.axvline(x=np.mean(amp_fast),color="gray",linewidth=3,linestyle="--")#,label="Fast (Pre-Transient, Abrupt Simulations)")
        p1 = plt.axvline(x=np.mean(amp_trns),color="red",linewidth=3,linestyle="-")#,label="Transient")
        d2 = sns.distplot(list(amp_eq),color="black",hist=False,kde=True,kde_kws={"linewidth":3,"alpha":1})#,label="Equilibrium")
        
    if i == 0:
        plt.ylabel("Density",labelpad=10)
    plt.yticks([])
    
    if i == 0:
        plt.xlim([0.4,1.55])
    elif i == 1:
        plt.xlim([0.3,1.0])
    else:
        plt.xlim([0.25,0.8])
        
    if i == 1:
        plt.xlabel("ENSO Amplitude ("+r"$\degree$C)")
        
    plt.title(ex[i],fontsize=28,pad=10)
    

plt.figtext(0.115,0.935,r'$\bf{a}$',fontsize=32)
plt.figtext(0.4,0.935,r'$\bf{b}$',fontsize=32)
plt.figtext(0.685,0.935,r'$\bf{c}$',fontsize=32)

plt.figtext(0.115,0.545,r'$\bf{d}$',fontsize=32)
plt.figtext(0.4,0.545,r'$\bf{e}$',fontsize=32)
plt.figtext(0.685,0.545,r'$\bf{f}$',fontsize=32)


plt.savefig("../Figures/Figure3.pdf")
plt.show()
```
