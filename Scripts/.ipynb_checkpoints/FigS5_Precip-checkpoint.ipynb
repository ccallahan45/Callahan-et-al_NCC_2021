Function for setting boxplot color

```python
def set_boxplot_col(b,n,c,c_double,lw,median):
    
    # This function sets the colors of a given boxplot b
    # It auto-sets the colors of the boxes, whiskers, caps, and fliers to c[m]
    # and it sets the colors of the medians if median=True
    # n is the number of instances of the boxplot and lw is the desired
    # linewidth
    # c should be a vector of colors
    
    for m in range(0,n*2,1):
        if m < n:
            b["boxes"][m].set(color=c[m],linewidth=lw)
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
            #b["fliers"][m].set(markerfacecolor=c[m],markeredgecolor=c[m])
            if median==True:
                b["medians"][m].set(color=c[m],linewidth=lw)
        else:
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
```

#### Analysis
Running mean length

```python
rm_yrs = 100
rm_length = rm_yrs*3
```

```python
# DJF nino3 rainfall greater than 5mm per day
```

```python
%%capture

control_count = np.zeros(len(modelnames_file))
transient_count = np.zeros(len(modelnames_file))
transient_change = np.zeros(len(modelnames_file))
equilibrium_count = np.zeros(len(modelnames_file))
equilibrium_change = []

for i in np.arange(0,len(modelnames_file),1):
    
    model, exp = modelnames_file[i].split("_")
    print(modelnames_file[i]) 
    
    precip_control = xr.DataArray(xr.open_dataset(loc_precip+"precip_nino3_"+model+"_control.nc").data_vars["pr"]) * 86400
    precip_exp = xr.DataArray(xr.open_dataset(loc_precip+"precip_nino3_"+model+"_"+exp+".nc").data_vars["pr"]) * 86400
    
    # limit to DJF
    precip_control_djf = precip_control[precip_control.time.dt.season=="DJF"]
    precip_exp_djf = precip_exp[precip_exp.time.dt.season=="DJF"]
    
    # Count > 5
    precip_control_bool = (precip_control_djf >= 5).astype(int)
    precip_exp_bool = (precip_exp_djf >= 5).astype(int)
    
    # Sum
    een_control = (precip_control_bool.sum(dim="time"))/(len(precip_control_bool)/3)
    control_count[i] = een_control.values
    
    if exp.startswith("abrupt"):
        precip_exp_transient = precip_exp_bool[(50*3):(150*3)]
        precip_exp_eq = precip_exp_bool[(150*3):]
    else:
        precip_exp_transient = precip_exp_bool[0:(140*3)]
        precip_exp_eq = precip_exp_bool[(140*3):]
    
    een_transient = (precip_exp_transient.sum(dim="time"))/(len(precip_exp_transient)/3)
    transient_count[i] = een_transient.values
    transient_change[i] = een_transient - een_control
    #transient_change[i] = ((een_transient - een_control)/een_control) * 100
    
    een_eq_rolling = precip_exp_eq.rolling(time=(100*3),center=True).sum()
    een_eq = een_eq_rolling[(50*3):(len(een_eq_rolling)-(50*3))]/100 #(100*3)
    equilibrium_count[i] = np.median(een_eq.values)
    equilibrium_change.append((een_eq - een_control).values)
    #equilibrium_change.append(((een_eq - een_control)/een_control) * 100)
```

Precip mean change

```python
%%capture

transient_change_mean = np.zeros(len(modelnames_file))
equilibrium_change_mean = []

for i in np.arange(0,len(modelnames_file),1):
    
    model, exp = modelnames_file[i].split("_")
    print(modelnames_file[i]) 
    
    precip_control = xr.DataArray(xr.open_dataset(loc_precip+"precip_nino3_"+model+"_control.nc").data_vars["pr"]) * 86400
    precip_exp = xr.DataArray(xr.open_dataset(loc_precip+"precip_nino3_"+model+"_"+exp+".nc").data_vars["pr"]) * 86400
    
    if exp.startswith("abrupt"):
        precip_exp_transient = precip_exp[(50*12):(150*12)]
        precip_exp_eq = precip_exp[(150*12):]
    else:
        precip_exp_transient = precip_exp[0:(140*12)]
        precip_exp_eq = precip_exp[(140*12):]
        
    # limit to DJF
    precip_control_djf = precip_control[precip_control.time.dt.season=="DJF"]
    precip_transient_djf = precip_exp_transient[precip_exp_transient.time.dt.season=="DJF"]
    precip_eq_djf = precip_exp_eq[precip_exp_eq.time.dt.season=="DJF"]
    
    # calculate change
    precip_control_mean = precip_control_djf.mean(dim="time").values 
    precip_transient_mean = precip_transient_djf.mean(dim="time").values
    
    een_eq_rolling = precip_eq_djf.rolling(time=(100*3),center=True).mean()
    een_eq = een_eq_rolling[(50*3):(len(een_eq_rolling)-(50*3))] #/(100*3)
    een_eq_change = ((een_eq - precip_control_mean)/precip_control_mean)*100
    #precip_eq_mean = precip_eq_djf.mean(dim="time").values
    
    transient_change_mean[i] = ((precip_transient_mean - precip_control_mean)/precip_control_mean)*100
    #equilibrium_change_mean[i] = ((precip_eq_mean - precip_control_mean)/precip_control_mean)*100
    equilibrium_change_mean.append(een_eq_change.values)
```

#### Plot

```python
fig = plt.figure(figsize=(23,16))

rcParams["font.family"] = "sans-serif"
rcParams["font.sans-serif"] = ["Helvetica Neue"]
rcParams["font.size"] = 30.0
rcParams["axes.linewidth"] = 1.5
rcParams['xtick.major.size'] = 8
rcParams['xtick.major.width'] = 1.5
rcParams['ytick.major.size'] = 8
rcParams['ytick.major.width'] = 1.5


gs1 = gridspec.GridSpec(2,1)
gs1.update(left=0.1,right=0.97,top=0.95,bottom=0.25,wspace=0.2,hspace=0.15)

for i in np.arange(0,2,1):

    ax = plt.subplot(gs1[i,0])
    
    if i == 0:
        eq = equilibrium_change_mean
        trns = transient_change_mean
    else:
        eq = equilibrium_change
        trns = transient_change
        
    
    b1 = plt.boxplot(eq,whis=[5,95],sym="")
    set_boxplot_col(b1,21,np.array(colors)/255.,np.array(colors_double)/255.,3,True)

    plt.scatter(np.arange(1,22,1),trns,c=np.array(colors)/255.,s=75,marker="d")
    plt.plot([1,2,3],[np.median(eq[0]),np.median(eq[1]),np.median(eq[2])],linewidth=2.5,color=np.array(colors[0])/255.)
    plt.plot([4,5,6],[np.median(eq[3]),np.median(eq[4]),np.median(eq[5])],linewidth=2.5,color=np.array(colors[3])/255.)
    plt.plot([12,13,14,15],[np.median(eq[11]),np.median(eq[12]),np.median(eq[13]),np.median(eq[14])],linewidth=2.5,color=np.array(colors[12])/255.)
    plt.plot([19,20,21],[np.median(eq[18]),np.median(eq[19]),np.median(eq[20])],linewidth=2.5,color=np.array(colors[18])/255.)
        
    #for i in np.arange(0,len(ex1)):
        #if pvals[i] >= 0.05:
            #plt.scatter(i+1,-75,color="white",edgecolors=[0.5,0.5,0.5],s=100)

    plt.axhline(y=0,linewidth=1.5,color="k",linestyle="-")

    if i == 1:
        plt.xticks(np.arange(1,22,1),modelnames_fig,rotation=45,horizontalalignment="right")
    else:
        plt.xticks([])
    
    plt.xlim([0,22])
    
    if i == 0:
        #plt.ylim([-60,40])
        plt.ylabel("$\Delta$ Precipitation (%)")
    else:
        #plt.ylim([-150,100])
        plt.ylabel("$\Delta$ Extreme El Nino Precip "+"\n"+"(# per yr)",ha="center")


plt.figtext(0.11,0.91,"a",fontsize=45)
plt.figtext(0.11,0.53,"b",fontsize=45)


plt.savefig("../Figures/FigS5_Precip.pdf")
plt.show()
```
