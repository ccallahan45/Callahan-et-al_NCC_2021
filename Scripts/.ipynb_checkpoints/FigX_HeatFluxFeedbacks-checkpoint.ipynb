Set boxplot color function

```python
def set_boxplot_col(b,n,c,c_double,lw,median):
    
    # This function sets the colors of a given boxplot b
    # It auto-sets the colors of the boxes, whiskers, caps, and fliers to c[m]
    # and it sets the colors of the medians if median=True
    # n is the number of instances of the boxplot and lw is the desired
    # linewidth
    # c should be a vector of colors
    
    for m in range(0,n*2,1):
        if m < n:
            b["boxes"][m].set(color=c[m],linewidth=lw)
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
            #b["fliers"][m].set(markerfacecolor=c[m],markeredgecolor=c[m])
            if median==True:
                b["medians"][m].set(color=c[m],linewidth=lw)
        else:
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
```

```python
for i in np.arange(0,len(modelnames_file_v2),1): 

    model, exp = modelnames_file_v2[i].split("_")
    print(modelnames_file_v2[i])
    
    if exp.startswith("contr"):
        hfss_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfss_nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["hfss"])
        hfls_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfls_nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["hfls"])
        hfnet_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfnet_nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["hfnet"])
        nino34_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["nino34"])
    else:
        hfss_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfss_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["hfss"])
        hfls_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfls_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["hfls"])
        hfnet_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfnet_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["hfnet"])
        nino34_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino34"])
        
        
    if model == "GISSE2R":
        hfss_eq = hfss_anom[len(hfss_anom)-(1000*12):]
        hfls_eq = hfss_anom[len(hfls_anom)-(1000*12):]
        hfnet_eq = hfnet_anom[len(hfnet_anom)-(1000*12):]
        nino34_eq = nino34_anom[len(nino34_anom)-(1000*12):]
    else: 
        if exp.startswith("abrupt"):
            #hfss_eq = hfss_anom[(150*12):(950*12)+1]
            #hfls_eq = hfls_anom[(150*12):(950*12)+1]
            #hfnet_eq = hfnet_anom[(150*12):(950*12)+1]
            #nino34_eq = nino34_anom[(150*12):(950*12)+1]
            hfss_eq = hfss_anom[(150*12):]
            hfls_eq = hfls_anom[(150*12):]
            hfnet_eq = hfnet_anom[(150*12):]
            nino34_eq = nino34_anom[(150*12):]
        elif exp.startswith("1pct"):
            hfss_eq = hfss_anom[(140*12):]
            hfls_eq = hfls_anom[(140*12):]
            hfnet_eq = hfnet_anom[(140*12):]
            nino34_eq = nino34_anom[(140*12):]
        else:
            hfss_eq = hfss_anom
            hfls_eq = hfls_anom
            hfnet_eq = hfnet_anom
            nino34_eq = nino34_anom
        
    ## rolling regressions
    
    heatflux_enso_exp = pd.DataFrame(np.transpose([hfss_eq.values,hfls_eq.values,hfnet_eq.values,nino34_eq.values]),
                                     columns=["hfss","hfls","hfnet","Nino34"])
    
    
    hfss_model = RollingOLS.from_formula("hfss ~ Nino34",data=heatflux_enso_exp,window=100*12).fit()
    hfls_model = RollingOLS.from_formula("hfls ~ Nino34",data=heatflux_enso_exp,window=100*12).fit()
    hfnet_model = RollingOLS.from_formula("hfnet ~ Nino34",data=heatflux_enso_exp,window=100*12).fit()
    
    ## get coefficients
    
    beta_hfss_exp = hfss_model.params["Nino34"].values[~np.isnan(hfss_model.params["Nino34"].values)]    
    beta_hfls_exp = hfls_model.params["Nino34"].values[~np.isnan(hfls_model.params["Nino34"].values)]
    beta_hfnet_exp = hfnet_model.params["Nino34"].values[~np.isnan(hfnet_model.params["Nino34"].values)]
    
    beta_hfss_all.append(beta_hfss_exp)
    beta_hfls_all.append(beta_hfls_exp)
    beta_hfnet_all.append(beta_hfnet_exp)
    
    #print("sensible heat flux feedback:")
    #print("control mean: "+str(np.mean(beta_hfss_ctrl))+", sd: "+str(np.std(beta_hfss_ctrl)))
    #print("forced mean: "+str(np.mean(beta_hfss_exp))+", sd: "+str(np.std(beta_hfss_exp)))
    
    #print("latent heat flux feedback:")
    #print("control mean: "+str(np.mean(beta_hfls_ctrl))+", sd: "+str(np.std(beta_hfls_ctrl)))
    #print("forced mean: "+str(np.mean(beta_hfls_exp))+", sd: "+str(np.std(beta_hfls_exp)))
    
    #print("total heat flux feedback:")
    #print("control mean: "+str(np.mean(beta_hfnet_ctrl))+", sd: "+str(np.std(beta_hfnet_ctrl)))
    #print("forced mean: "+str(np.mean(beta_hfnet_exp))+", sd: "+str(np.std(beta_hfnet_exp)))
    
    #if i == 0:
        #beta_hfss_all = np.full((len(modelnames_file_v2),len(beta_hfss_exp)),np.nan)
        #beta_hfls_all = np.full((len(modelnames_file_v2),len(beta_hfls_exp)),np.nan)
        #beta_hfnet_all = np.full((len(modelnames_file_v2),len(beta_hfnet_exp)),np.nan)

    #beta_hfss_exp_all[i,:] = beta_hfss_exp
    #beta_hfss_ctrl_all[i,:] = beta_hfss_ctrl
    #beta_hfls_all[i,:] = beta_hfls_exp
    #beta_hfnet_all[i,:] = beta_hfnet_exp
```

```python
beta_hfls_change = []
beta_hfls_exps = []
beta_hfls_ctrls = []
for i in np.arange(0,len(modelnames_file),1): 

    model, exp = modelnames_file[i].split("_")
    print(modelnames_file[i])
    
    hfls_control_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfls_nino34_"+model+"_control_anom_detrend.nc").data_vars["hfls"])
    nino34_control_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_control_anom_detrend.nc").data_vars["nino34"])

    hfls_exp_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfls_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["hfls"])
    nino34_exp_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino34"])
        
        
    if model == "GISSE2R":
        hfls_ctrl = hfls_control_anom[len(hfls_control_anom)-(1000*12):]
        nino34_ctrl = nino34_control_anom[len(nino34_control_anom)-(1000*12):]
        hfls_eq = hfls_exp_anom[len(hfls_exp_anom)-(1000*12):]
        nino34_eq = nino34_exp_anom[len(nino34_exp_anom)-(1000*12):]
    else:
        if exp.startswith("abrupt"):
            hfls_eq = hfls_exp_anom[(150*12):]
            nino34_eq = nino34_exp_anom[(150*12):]
            hfls_ctrl = hfls_control_anom
            nino34_ctrl = nino34_control_anom
        elif exp.startswith("1pct"):
            hfls_eq = hfls_exp_anom[(140*12):]
            nino34_eq = nino34_exp_anom[(140*12):]
            hfls_ctrl = hfls_control_anom
            nino34_ctrl = nino34_control_anom
        
        
    ## rolling regressions
    
    heatflux_enso_exp = pd.DataFrame(np.transpose([hfls_eq.values,nino34_eq.values]),
                                     columns=["hfls","Nino34"])
    heatflux_enso_control = pd.DataFrame(np.transpose([hfls_ctrl.values,nino34_ctrl.values]),
                                     columns=["hfls","Nino34"])
    
    hfls_model_exp = RollingOLS.from_formula("hfls ~ Nino34",data=heatflux_enso_exp,window=100*12).fit()
    hfls_model_ctrl = RollingOLS.from_formula("hfls ~ Nino34",data=heatflux_enso_control,window=100*12).fit()
    
    
    ## get coefficients
    
    beta_hfls_exp = hfls_model_exp.params["Nino34"].values[~np.isnan(hfls_model_exp.params["Nino34"].values)]
    beta_hfls_ctrl = hfls_model_ctrl.params["Nino34"].values[~np.isnan(hfls_model_ctrl.params["Nino34"].values)]
    
    beta_hfls_change_model = beta_hfls_exp - np.mean(beta_hfls_ctrl)
    beta_hfls_exps.append(beta_hfls_exp)
    beta_hfls_ctrls.append(beta_hfls_ctrl)
    beta_hfls_change.append(beta_hfls_change_model)
```
