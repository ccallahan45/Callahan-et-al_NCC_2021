This notebook plots one of the supplementary figures for Callahan et al.,robust long-term decline in ENSO amplitude. It shows how the control and forced zonal temperature gradients in each LongRunMIP model compare to the gradient during typical El Nino events.

#### Mechanics
Read in dependencies

```python
import xarray as xr
import numpy as np
import sys
import os
import datetime
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap, cm
from matplotlib import rcParams
import matplotlib.gridspec as gridspec
import seaborn as sns
from scipy.io import loadmat
from scipy import signal
```

Set data location

```python
loc_nino34 = "../Data/ENSO_Indices/"
loc_ztg = "../Data/ZTG/"
```

Set models list, experiments, and colors

```python
modelnames_fig = ['CCSM3 abrupt 2x','CCSM3 abrupt 4x','CCSM3 abrupt 8x', \
    'CESM1.0.4 abrupt 2x','CESM1.0.4 abrupt 4x','CESM1.0.4 abrupt 8x', 'CNRM-CM6.1 abrupt4x', \
    'GFDL-CM3 1pct 2x','GFDL-ESM2M 1pct 2x','GISS-E2-R 1pct 4x', \
    'GISS-E2-R abrupt 4x','HadCM3L abrupt 2x','HadCM3L abrupt 4x', \
    'HadCM3L abrupt 6x','HadCM3L abrupt 8x','IPSL-CM5A-LR abrupt 4x', \
    'MIROC3.2 1pct 2x','MIROC3.2 1pct 4x','MPIESM-1.2 abrupt 2x', \
    'MPIESM-1.2 abrupt 4x','MPIESM-1.2 abrupt 8x'] 

modelnames_file = ['CCSM3_abrupt2x','CCSM3_abrupt4x','CCSM3_abrupt8x', \
    'CESM104_abrupt2x','CESM104_abrupt4x','CESM104_abrupt8x', \
    'CNRMCM61_abrupt4x','GFDLCM3_1pct2x','GFDLESM2M_1pct2x','GISSE2R_1pct4x', \
    'GISSE2R_abrupt4x','HadCM3L_abrupt2x','HadCM3L_abrupt4x', \
    'HadCM3L_abrupt6x','HadCM3L_abrupt8x','IPSLCM5A_abrupt4x', \
    'MIROC32_1pct2x','MIROC32_1pct4x','MPIESM12_abrupt2x', \
    'MPIESM12_abrupt4x','MPIESM12_abrupt8x'] 

runtype = ['abrupt','abrupt','abrupt','abrupt','abrupt','abrupt','lin','lin','lin', \
            'abrupt','abrupt','abrupt','abrupt','abrupt','abrupt', \
            'lin','lin','abrupt','abrupt','abrupt','abrupt']

colors = [[0,238,0],[0,238,0],[0,238,0], \
          [34,139,34],[34,139,34],[34,139,34],[135,206,255],[16,78,139],[30,144,255], \
          [255,110,180],[255,110,180],[255,0,0],[255,0,0],[255,0,0],[255,0,0], \
          [255,193,37],[122,55,139],[122,55,139],[153,153,153],[153,153,153],[153,153,153]]

colors_double = []
for i in np.arange(0,len(colors),1):
    colors_double.append(colors[i])
    colors_double.append(colors[i])
```

Function for hiding right and top axis

```python
def hide_right_and_top(axis):
    
    # This function hides the right and top axes
    # of a given axis object
    # For purely aesthetic purposes
    
    # Hide the right and top spines
    axis.spines['right'].set_visible(False)
    axis.spines['top'].set_visible(False)

    # Only show ticks on the left and bottom spines
    axis.yaxis.set_ticks_position('left')
    axis.xaxis.set_ticks_position('bottom')
```

#### Analysis

Create empty variables

```python
ztg_change = np.zeros(len(modelnames_fig))
ztg_thresh = np.zeros(len(modelnames_fig))
ztg_new = np.zeros(len(modelnames_fig))
ztg_old = np.zeros(len(modelnames_fig))
```

Loop through models

```python
for i in np.arange(0,len(modelnames_file),1):
    
    model, exp = modelnames_file[i].split("_")
    print(model)
    
    # Find control and forced zonal temperature gradients
    
    if ((model == "MIROC32") | (model == "MPIESM12")):
        ztg_c = xr.DataArray(xr.open_dataset(loc_ztg+"ztg_"+model+"_control.nc",decode_times=False).data_vars["ztg"])
        ztg_f = xr.DataArray(xr.open_dataset(loc_ztg+"ztg_"+model+"_"+exp+".nc",decode_times=False).data_vars["ztg"])
    else:
        ztg_c = xr.DataArray(xr.open_dataset(loc_ztg+"ztg_"+model+"_control.nc").data_vars["ztg"])
        ztg_f = xr.DataArray(xr.open_dataset(loc_ztg+"ztg_"+model+"_"+exp+".nc").data_vars["ztg"])
        
    if runtype[i] == "abrupt":
        nyear_e = 150
    else:
        nyear_e = 140
    
    ztg_e = ztg_f[(nyear_e*12)-1:(ztg_f.shape[0]-1)]
    ztg_old[i] = np.mean(ztg_c)
    ztg_new[i] = np.mean(ztg_e)
    
    # Now define ENSO-based ZTG threshold
    
    nino34_c = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_control_anom_detrend2.nc").data_vars["nino34"])
    
    ctrl_en1 = np.percentile(nino34_c.values,75)
    ctrl_en2 = np.percentile(nino34_c.values,90)
    
    ztg_threshold = np.mean(ztg_c[(nino34_c.values >= ctrl_en1) & (nino34_c.values <= ctrl_en2)])
    ztg_thresh[i] = ztg_threshold
```

#### Plot

```python
fig = plt.figure(figsize=(22,11))

rcParams["font.family"] = "sans-serif"
rcParams["font.sans-serif"] = ["Helvetica Neue"]
rcParams["font.size"] = 25.0
rcParams["axes.linewidth"] = 2
rcParams['xtick.major.size'] = 8
rcParams['xtick.major.width'] = 2
rcParams['ytick.major.size'] = 8
rcParams['ytick.major.width'] = 2

gs1 = gridspec.GridSpec(1,1)
gs1.update(left=0.115,right=0.975,top=0.93,bottom=0.37,wspace=0.2,hspace=0.28)

ax = plt.subplot(gs1[0,0])
hide_right_and_top(ax)

s = 1000

for i in np.arange(0,len(modelnames_file),1):

    cs = list(np.array(colors[i])/255.)
    if i == len(modelnames_file)-1:
        p1 = plt.scatter(i,ztg_old[i]-ztg_thresh[i],s,edgecolors=cs,facecolor="white",linewidth=1.5,label="Control mean gradient")
        p2 = plt.scatter(i,ztg_new[i]-ztg_thresh[i],s,edgecolors=cs,facecolor=cs,label="High-CO$_2$ mean gradient")
    else:
        p1 = plt.scatter(i,ztg_old[i]-ztg_thresh[i],s,edgecolors=cs,linewidth=1.5,facecolor="white")
        p2 = plt.scatter(i,ztg_new[i]-ztg_thresh[i],s,edgecolors=cs,facecolor=cs)
        
    
plt.xlim([-0.5,20.5])
plt.ylim([-0.8,0.8])
plt.yticks([-0.8,-0.4,0,0.4,0.8])
plt.axhline(y=0,linewidth=2,linestyle="--",color="black")
plt.ylabel("Difference from gradient during"+"\n"+ "El Nino in control ($\degree$C)")
plt.xticks(ticks=np.arange(0,len(modelnames_fig),1),labels=modelnames_fig,rotation=45,horizontalalignment="right")

plt.legend(loc="upper left",frameon=False,scatteryoffsets=[0.5],markerscale=0.8,handletextpad=0.1,
          bbox_to_anchor=(0,1.05))

#plt.title("ENSO-like mean gradients")

plt.savefig("../Figures/FigS10_Permanent_ElNino.pdf")
plt.show()
```
