This notebook plots a supplementary figure for Callahan et al., Robust long-term decline in ENSO amplitude. Specifically, it plots longitudinal temperature profiles for each model to demonstrate changes in zonal temperature gradients across those models.

#### Mechanics
Read in dependencies

```python
import xarray as xr
import numpy as np
import sys
import os
import datetime
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap, cm
from matplotlib import rcParams
import matplotlib.gridspec as gridspec
import seaborn as sns
from scipy.io import loadmat
from scipy import signal
```

Set data location

```python
loc_in = "../Data/Pacific_Profiles/"
loc_nino34 = "../Data/ENSO_Indices/"
```

Set models list and experiments

```python
modelnames_fig = ['CCSM3 abrupt 2x','CCSM3 abrupt 4x','CCSM3 abrupt 8x', \
    'CESM1.0.4 abrupt 2x','CESM1.0.4 abrupt 4x','CESM1.0.4 abrupt 8x', 'CNRM-CM6.1 abrupt4x', \
    'GFDL-CM3 1pct 2x','GFDL-ESM2M 1pct 2x','GISS-E2-R 1pct 4x', \
    'GISS-E2-R abrupt 4x','HadCM3L abrupt 2x','HadCM3L abrupt 4x', \
    'HadCM3L abrupt 6x','HadCM3L abrupt 8x','IPSL-CM5A-LR abrupt 4x', \
    'MIROC3.2 1pct 2x','MIROC3.2 1pct 4x','MPIESM-1.2 abrupt 2x', \
    'MPIESM-1.2 abrupt 4x','MPIESM-1.2 abrupt 8x']

modelnames_file = ['CCSM3_abrupt2x','CCSM3_abrupt4x','CCSM3_abrupt8x', \
    'CESM104_abrupt2x','CESM104_abrupt4x','CESM104_abrupt8x', \
    'CNRMCM61_abrupt4x','GFDLCM3_1pct2x','GFDLESM2M_1pct2x','GISSE2R_1pct4x', \
    'GISSE2R_abrupt4x','HadCM3L_abrupt2x','HadCM3L_abrupt4x', \
    'HadCM3L_abrupt6x','HadCM3L_abrupt8x','IPSLCM5A_abrupt4x', \
    'MIROC32_1pct2x','MIROC32_1pct4x','MPIESM12_abrupt2x', \
    'MPIESM12_abrupt4x','MPIESM12_abrupt8x']

runtype = ['abrupt','abrupt','abrupt','abrupt','abrupt','abrupt','lin','lin','lin', \
            'abrupt','abrupt','abrupt','abrupt','abrupt','abrupt', \
            'lin','lin','abrupt','abrupt','abrupt','abrupt']
```

Function for hiding top axis

```python
def hide_top(axis):
    
    # This function hides the top axis
    # of a given axis object
    # For purely aesthetic purposes
    
    # Hide the top spine
    axis.spines['top'].set_visible(False)

    # Only show ticks on the bottom spine
    axis.xaxis.set_ticks_position('bottom')
```

#### Analysis

We'll have to create the figure right off the bat, because we're going to loop inside the figure and plot each model independently

```python
fig = plt.figure(figsize=(36,26))

rcParams["font.family"] = "sans-serif"
rcParams["font.sans-serif"] = ["Helvetica Neue"]
rcParams["font.size"] = 19.0
rcParams["axes.linewidth"] = 2
rcParams['xtick.major.size'] = 8
rcParams['xtick.major.width'] = 2
rcParams['ytick.major.size'] = 8
rcParams['ytick.major.width'] = 2

gs1 = gridspec.GridSpec(4,6)
gs1.update(left=0.05,right=0.91,top=0.93,bottom=0.1,wspace=0.53,hspace=0.42)

for i in np.arange(0,len(modelnames_file),1):
        
    model, exp = modelnames_file[i].split("_")
    print(modelnames_file[i])
    
    if ((model == "CCSM3") | (model == "CESM104")):
        ax = plt.subplot(gs1[0,i])
    elif ((model == "GFDLCM3") | (model == "GFDLESM2M") | (model == "GISSE2R") | (model == "CNRMCM61")):
        ax = plt.subplot(gs1[1,i-6])
    elif ((model == "HadCM3L") | (model == "IPSLCM5A")):
        ax = plt.subplot(gs1[2,i-11])
    else:
        ax = plt.subplot(gs1[3,i-16])
        
    #ax = plt.subplot(gs1[i])
    
    hide_top(ax)
    
    f_in_profile_f = xr.open_dataset(loc_in+"pacific_lon_profile_"+model+"_"+exp+".nc")
    f_in_profile_c = xr.open_dataset(loc_in+"pacific_lon_profile_"+model+"_control.nc")

    # Read in variables

    profile_f = xr.DataArray(f_in_profile_f.data_vars["profile"])
    profile_c = xr.DataArray(f_in_profile_c.data_vars["profile"])
    
    nino34_c = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_control_anom_detrend2.nc").data_vars["nino34"])
    nino34_f = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino34"])
    
    profile_lons = profile_f.coords["lon"].values

    ind170 = int(np.argmin(np.abs(profile_lons - 170)))
    ind215 = int(np.argmin(np.abs(profile_lons - 215)))
    ind260 = int(np.argmin(np.abs(profile_lons - 260)))
    
    # Find thresholds for La Nina and El Nino

    ctrl_en1 = np.percentile(nino34_c.values,75)
    ctrl_en2 = np.percentile(nino34_c.values,90)
    ctrl_ln1 = np.percentile(nino34_c.values,10)
    ctrl_ln2 = np.percentile(nino34_c.values,25)

    f_en1 = np.percentile(nino34_f.values,75)
    f_en2 = np.percentile(nino34_f.values,90)
    f_ln1 = np.percentile(nino34_f.values,10)
    f_ln2 = np.percentile(nino34_f.values,25)

    # Find average longitudinal profiles

    pc = profile_c.values
    profile_ctrl = np.mean(pc,axis=0)
    profile_ctrl_en = np.mean(pc[(nino34_c.values >= ctrl_en1) & (nino34_c.values <= ctrl_en2),:],axis=0)
    profile_ctrl_ln = np.mean(pc[(nino34_c.values >= ctrl_ln1) & (nino34_c.values <= ctrl_ln2),:],axis=0)

    if runtype[i] == "abrupt":
        nyear_e = 150
    else:
        nyear_e = 140

    profile_e = profile_f[nyear_e*12-1:profile_f.shape[0]-1,:]
    nino34_e = nino34_f[nyear_e*12-1:nino34_f.shape[0]-1]

    pe = profile_e.values
    profile_eq = np.mean(pe,axis=0)
    profile_eq_en = np.mean(pe[(nino34_e.values >= f_en1) & (nino34_e.values <= f_en2),:],axis=0)
    profile_eq_ln = np.mean(pe[(nino34_e.values >= f_ln1) & (nino34_e.values <= f_ln2),:],axis=0)
    
    # Figure out the y-axis scale
    
    if model == "CCSM3":
        scaling = 5
    elif model == "CESM104":
        scaling = 6
    elif (model == "GFDLCM3") | (model == "GFDLESM2M"):
        scaling = 6
    elif (model == "GISSE2R"):
        scaling = 4
    elif model == "HadCM3L":
        scaling = 5
    elif model == "IPSLCM5A":
        scaling = 5
    elif model == "MIROC32":
        scaling = 5
    elif (model == "MPIESM12") | (model == "MPIESM11"):
        scaling = 6
    elif model == "CNRMCM61":
        scaling = 5
    #scaling = 6
    
    # Plot
    
    p1 = plt.plot(profile_ctrl,linewidth=3,linestyle="-",label="Control mean gradient",color="k")
    p2 = plt.plot(profile_ctrl_en,linewidth=3,linestyle="--",label="Control EN gradient",color="k")
    p3 = plt.plot(profile_ctrl_ln,linewidth=3,linestyle="-.",label="Control LN gradient",color="k")

    min_ctrl = int(np.min(profile_ctrl_ln[ind170:ind260]))
    plt.ylabel("Control temperature (K)",color="black")
    plt.ylim([min_ctrl,min_ctrl+scaling])
    #plt.ylim([298,303])
    plt.xlabel("Longitude")

    if modelnames_file[i] == "GISSE2R_abrupt4x":
        plt.legend(loc="upper right",frameon=False,fontsize=30,bbox_to_anchor=(3, 0.5))

    # Forced
    ax1 = ax.twinx()
    hide_top(ax1)
    axcol = "red"

    p1 = plt.plot(profile_eq,linewidth=3,linestyle="-",label="High-CO$_2$ mean gradient",color="r")
    p2 = plt.plot(profile_eq_en,linewidth=3,linestyle="--",label="High-CO$_2$ EN gradient",color="r")
    p3 = plt.plot(profile_eq_ln,linewidth=3,linestyle="-.",label="High-CO$_2$ LN gradient",color="r")
    
    min_eq = int(np.min(profile_eq_ln[ind170:ind260]))
    plt.ylim([min_eq,min_eq+scaling])
    #plt.ylim([301,306])
    plt.ylabel("Equilibrium temperature (K)",color=axcol)
    ax1.tick_params(axis="y",labelcolor=axcol)

    
    plt.xticks(ticks=[ind170,ind215,ind260],labels=["170 E","215 E","260 E"])
    plt.xlim([int(ind170)-2,int(ind260)+2])
    
    if modelnames_file[i] == "GISSE2R_abrupt4x":
        plt.legend(loc="upper right",frameon=False,fontsize=30,bbox_to_anchor=(3.14, 1.1))

    plt.title(modelnames_fig[i])


    
plt.savefig("../Figures/FigS13_Pacific_Profiles.pdf")
plt.show()
```
