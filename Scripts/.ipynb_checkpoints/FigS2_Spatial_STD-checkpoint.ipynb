Custom detrending function

```python
def custom_detrend(data,order):
    
    # only for one-dimensional timeseries
    # numpy required
    
    x = np.arange(1,len(data)+1,1)
    
    model = np.polyfit(x,data,order)
    predicted = np.polyval(model,x)
    new = data - predicted
    
    return(new)
```

```python
def list_files(directory, extension):
    
    # Credit to: http://www.martinbroadhurst.com/listing-all-files-in-a-directory-with-a-certain-extension-in-python.html
    # Requires os.listdir
    
    return (f for f in os.listdir(directory) if f.endswith(extension))
```

Running mean

```python
def runmean(data,length):
    
    # This function takes in a numpy array of data (one-dimensional)
    # and takes the running mean according to the window length prescribed
    # in the function call
    # You need pandas for this function to work
    # It outputs a numpy array of the data

    data_rm = (pd.Series(data).rolling(window=length,min_periods=1,center=True).mean()).values
    return(data_rm)
```

Pattern correlation function

```python
# PATTERN CORRELATION
# Christopher Callahan
# April 2019
# Christopher.W.Callahan.GR@dartmouth.edu
# Based on NCL's pattern correlation function by Rob Nichols and Dennis Shea:
# https://www.ncl.ucar.edu/Document/Functions/Contributed/pattern_cor.shtml

# This function calculates the centered or uncentered
# pattern correlation between two arrays of two dimensions
# lat x lon

# centered pattern correlation means the values are converted
# to anomalies relative to an area-weighted mean before
# computing the correlation

# uncentered pattern correlation means the input values are
# left alone before the pattern correlation is computed

# the function takes inputs of x and y (the two grids)
# wgt for the area-weights, and centered as 0 for uncentered
# and 1 for centered

# This function assumes that the inputs are given as xarray DataArrays
# or numpy arrays

# finally, this function assumes numpy is installed


def pattern_cor(x, y, wgt, centered):

    import numpy as np

    #print("Computing pattern correlation using")
    #print("/dartfs-hpc/rc/lab/C/CMIG/ccallahan/INVERSIONS/FUNCTIONS/PATTERN_CORRELATION.py")

    # Check for input data errors

    if ((np.amin(x)==0) & (np.amax(x)==0)):
        print("PATTERN_COR FAILED: All values of first input grid are 0")
        return

    if ((np.amin(y)==0) & (np.amax(y)==0)):
        print("PATTERN_COR FAILED: All values of second input grid are 0")
        return

    if (len(x.shape) != 2):
        print("PATTERN_COR FAILED: First input grid must have 2 dimensions, not "+str(x.shape))
        return

    if (len(y.shape) != 2):
        print("PATTERN_COR FAILED: Second input grid must have 2 dimensions, not "+str(y.shape))
        return

    if (y.shape != x.shape):
        print("PATTERN_COR FAILED: First input grid has shape "+str(x.shape)+" and second input grid has shape "+str(y.shape))
        return

    # Compute pattern correlation

    if centered == 1: # centered pattern correlation

        sumwgt = np.nansum(wgt)
        xAvgArea = (np.nansum(x*wgt))/np.nansum(wgt)
        yAvgArea = (np.nansum(y*wgt))/np.nansum(wgt)

        xAnom = x - xAvgArea
        yAnom = y - yAvgArea

        xyCov = np.nansum(wgt*xAnom*yAnom)
        xAnom2 = np.nansum(wgt*(np.square(xAnom)))
        yAnom2 = np.nansum(wgt*(np.square(yAnom)))

    else: # uncentered pattern correlation

        xyCov = np.nansum(wgt*x*y)
        xAnom2 = np.nansum(wgt*(np.square(x)))
        yAnom2 = np.nansum(wgt*(np.square(y)))

    r = xyCov/(np.sqrt(xAnom2)*np.sqrt(yAnom2))

    return(r)
```

```python
if calculate:
    for i in np.arange(0,len(modelnames_file),1):

        model, exp = modelnames_file[i].split("_")
        print(modelnames_file[i])

        fname_exp_start = "tas_mon_"+modelnames_file[i]
        fname_control_start = "tas_mon_"+model+"_control"

        fname_exp = [f for f in os.listdir(loc_tas) if f.startswith(fname_exp_start)][0]
        fname_control = [f for f in os.listdir(loc_tas) if f.startswith(fname_control_start)][0]

        n1, n2, n3, n4, year_exp_nc = fname_exp.split("_")
        n1, n2, n3, n4, year_control_nc = fname_control.split("_")
        year_exp, nc = year_exp_nc.split(".")
        year_control, nc = year_control_nc.split(".")

        model_decode_times_list = ["MIROC32","MPIESM12"]

        if model in model_decode_times_list:
            tas_exp_global = xr.DataArray(xr.open_dataset(loc_tas+fname_exp,decode_times=False).data_vars["tas"]) 
            tas_control_global = xr.DataArray(xr.open_dataset(loc_tas+fname_control,decode_times=False).data_vars["tas"])
        else:
            tas_exp_global = xr.DataArray(xr.open_dataset(loc_tas+fname_exp).data_vars["tas"])
            tas_control_global = xr.DataArray(xr.open_dataset(loc_tas+fname_control).data_vars["tas"])

        if model == "HadCM3L":
            lat_tas = tas_exp_global.coords["latitude_1"]
            lon_tas = tas_exp_global.coords["longitude_1"]
            latname = "latitude_1"
            lonname = "longitude_1"
        else:
            lat_tas = tas_exp_global.coords["lat"]
            lon_tas = tas_exp_global.coords["lon"]
            latname = "lat"
            lonname = "lon"

        if lat_tas[0] > lat_tas[len(lat_tas.values)-1]:
            latmax = lat_min
            latmin = lat_max
        else:
            latmin = lat_min
            latmax = lat_max

        if exp.startswith("abrupt"):
            eq_period = 150
        else:
            eq_period = 140

        tas_exp_pacific = tas_exp_global[(eq_period*12):,:,:].loc[:,latmin:latmax,lon_min:lon_max]
        lat_pacific = tas_exp_pacific.coords[latname]
        lon_pacific = tas_exp_pacific.coords[lonname]
        time_exp = tas_exp_pacific.coords["time"]
        tas_control_pacific = tas_control_global.loc[:,latmin:latmax,lon_min:lon_max]
        time_control = tas_control_pacific.coords["time"]

        tas_exp_pacific_dt = xr.DataArray(np.full((tas_exp_pacific.values.shape),np.nan),
                                          coords=[time_exp,lat_pacific,lon_pacific],
                                          dims=["time","lat","lon"])
        tas_control_pacific_dt = xr.DataArray(np.full((tas_control_pacific.values.shape),np.nan),
                                          coords=[time_control,lat_pacific,lon_pacific],
                                          dims=["time","lat","lon"])
        for j in np.arange(0,len(lat_pacific),1):
            for k in np.arange(0,len(lon_pacific),1):
                tas_exp_pacific_dt[:,j,k] = runmean(custom_detrend(tas_exp_pacific[:,j,k].values,ordr),12)
                tas_control_pacific_dt[:,j,k] = runmean(custom_detrend(tas_control_pacific[:,j,k].values,ordr),12)

        tas_exp_pacific_std = tas_exp_pacific_dt.std(dim="time")
        tas_control_pacific_std = tas_control_pacific_dt.std(dim="time")

        tas_exp_pacific_std_centered = tas_exp_pacific_std - tas_exp_pacific_std.mean(dim=["lat","lon"])
        tas_control_pacific_std_centered = tas_control_pacific_std - tas_control_pacific_std.mean(dim=["lat","lon"])

        #patcors[i] = pattern_cor(pacific_composite_f.values,pacific_composite_c.values,1.0,0)
        patcor = pattern_cor(tas_exp_pacific_std_centered.values,tas_control_pacific_std_centered.values,1.0,0)

        print("patcor with annual smoothing: "+str(patcor))



        # write out experiment
        tas_exp_pacific_std.name = "std"
        tas_exp_pacific_std.attrs["creation_date"] = str(datetime.datetime.now())
        tas_exp_pacific_std.attrs["created_by"] = "Christopher Callahan, Christopher.W.Callahan.GR@dartmouth.edu"
        tas_exp_pacific_std.attrs["data_description"] = "Pacific temperature standard deviation from tas, annual smoothing, detrended using order in file name"
        tas_exp_pacific_std.attrs["created_from"] = "FigS2_Spatial_STD.ipynb"

        fname_out_exp = loc_spatial_std+"tas_pacific_"+model+"_"+exp+"_std_detrend"+str(ordr)+"_v3.nc"
        tas_exp_pacific_std.to_netcdf(fname_out_exp,mode="w")
        print(fname_out_exp)


        # write out  control
        tas_control_pacific_std.name = "std"
        tas_control_pacific_std.attrs["creation_date"] = str(datetime.datetime.now())
        tas_control_pacific_std.attrs["created_by"] = "Christopher Callahan, Christopher.W.Callahan.GR@dartmouth.edu"
        tas_control_pacific_std.attrs["data_description"] = "Pacific temperature standard deviation from tas, annual smoothing, detrended using order in file name"
        tas_control_pacific_std.attrs["created_from"] = "FigS2_Spatial_STD.ipynb"

        fname_out_control = loc_spatial_std+"tas_pacific_"+model+"_control_std_detrend"+str(ordr)+"_v3.nc"
        tas_control_pacific_std.to_netcdf(fname_out_control,mode="w")
        print(fname_out_control)


        ## Calculate tas composites during non-el nino events defined using nino34

        nino34_ctrl_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_control_anom_detrend2.nc").data_vars["nino34"])
        nino34_forced_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino34"])

        nino34_eq = nino34_forced_anom[(eq_period*12):]
        en_eq = np.percentile(nino34_eq.values,75)
        en_ctrl = np.percentile(nino34_ctrl_anom.values,75)

        tas_exp_pacific_dt_anom = tas_exp_pacific_dt - np.nanmean(tas_exp_pacific_dt.values)
        tas_control_pacific_dt_anom = tas_control_pacific_dt - np.nanmean(tas_control_pacific_dt.values)

        tas_exp_pacific_composite_nonnino = tas_exp_pacific_dt_anom[(nino34_eq.values<en_eq)&(nino34_eq.values>np.median(nino34_eq.values)),:,:].mean(dim="time")
        tas_ctrl_pacific_composite_nonnino = tas_control_pacific_dt_anom[(nino34_ctrl_anom.values<en_ctrl)&(nino34_ctrl_anom.values>np.median(nino34_ctrl_anom.values)),:,:].mean(dim="time")


        # write out experiment
        tas_exp_pacific_composite_nonnino.name = "std"
        tas_exp_pacific_composite_nonnino.attrs["creation_date"] = str(datetime.datetime.now())
        tas_exp_pacific_composite_nonnino.attrs["created_by"] = "Christopher Callahan, Christopher.W.Callahan.GR@dartmouth.edu"
        tas_exp_pacific_composite_nonnino.attrs["data_description"] = "Composite of detrended temperatures during non-el nino events defined as nino3.4>p75"
        tas_exp_pacific_composite_nonnino.attrs["created_from"] = "FigS2_Spatial_STD.ipynb"

        fname_out_exp = loc_spatial_std+"tas_pacific_composite_"+model+"_"+exp+"_non_elnino_detrend"+str(ordr)+"_v4.nc"
        tas_exp_pacific_composite_nonnino.to_netcdf(fname_out_exp,mode="w")
        print(fname_out_exp)


        # write out  control
        tas_ctrl_pacific_composite_nonnino.name = "std"
        tas_ctrl_pacific_composite_nonnino.attrs["creation_date"] = str(datetime.datetime.now())
        tas_ctrl_pacific_composite_nonnino.attrs["created_by"] = "Christopher Callahan, Christopher.W.Callahan.GR@dartmouth.edu"
        tas_ctrl_pacific_composite_nonnino.attrs["data_description"] = "Composite of detrended temperatures during non-el nino events defined as nino3.4>p75"
        tas_ctrl_pacific_composite_nonnino.attrs["created_from"] = "FigS2_Spatial_STD.ipynb"

        fname_out_control = loc_spatial_std+"tas_pacific_composite_"+model+"_control_non_elnino_detrend"+str(ordr)+"_v4.nc"
        tas_ctrl_pacific_composite_nonnino.to_netcdf(fname_out_control,mode="w")
        print(fname_out_control)
```
