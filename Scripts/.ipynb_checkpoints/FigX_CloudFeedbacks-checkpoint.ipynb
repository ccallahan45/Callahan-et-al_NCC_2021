Boxplot color function

```python
def set_boxplot_col(b,n,c,c_double,lw,median):
    
    # This function sets the colors of a given boxplot b
    # It auto-sets the colors of the boxes, whiskers, caps, and fliers to c[m]
    # and it sets the colors of the medians if median=True
    # n is the number of instances of the boxplot and lw is the desired
    # linewidth
    # c should be a vector of colors
    
    for m in range(0,n*2,1):
        if m < n:
            b["boxes"][m].set(color=c[m],linewidth=lw)
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
            #b["fliers"][m].set(markerfacecolor=c[m],markeredgecolor=c[m])
            if median==True:
                b["medians"][m].set(color=c[m],linewidth=lw)
        else:
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
```

```python
lat_min = -5
lat_max = 5
lon_min = 190 # 210 #190
lon_max = 240 # 270 #240
```

```python
beta_rlut_all = []
beta_swcf_all = []
beta_rsds_all = []
beta_rsus_all = []
beta_rlds_all = []
beta_rlus_all = []
beta_swnet_all = []
beta_lwnet_all = []
beta_netsurf_all = []
beta_netflux_all = []

for i in np.arange(0,len(modelnames_file_v2),1):

    model, exp = modelnames_file_v2[i].split("_")
    print(modelnames_file_v2[i])
    
    if exp.startswith("contr"):
        rlut_anom = xr.DataArray(xr.open_dataset(loc_clouds+"rlut_nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["rlut"])
        swcf_anom = xr.DataArray(xr.open_dataset(loc_clouds+"swcf_nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["swcf"])
        nino34_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["nino34"])
        
        rsds_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rsds_nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["rsds"])
        rsus_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rsus_nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["rsus"])
        rlds_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rlds_nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["rlds"])
        rlus_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rlus_nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["rlus"])
        
        hfnet_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfnet_nino34_"+model+"_"+exp+"_anom_detrend.nc").data_vars["hfnet"])
        
    else:
        rlut_anom = xr.DataArray(xr.open_dataset(loc_clouds+"rlut_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["rlut"])
        swcf_anom = xr.DataArray(xr.open_dataset(loc_clouds+"swcf_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["swcf"])
        nino34_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino34"])
        
        rsds_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rsds_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["rsds"])
        rsus_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rsus_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["rsus"])
        rlds_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rlds_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["rlds"])
        rlus_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rlus_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["rlus"])
        hfnet_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfnet_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["hfnet"])
    
    if model == "GISSE2R":
        rlut_eq = rlut_anom[len(rlut_anom)-(1000*12):]
        swcf_eq = swcf_anom[len(swcf_anom)-(1000*12):]
        nino34_eq = nino34_anom[len(nino34_anom)-(1000*12):]
        
        rsds_eq = rsds_anom[len(rsds_anom)-(1000*12):]
        rsus_eq = rsus_anom[len(rsus_anom)-(1000*12):]*-1
        rlds_eq = rlds_anom[len(rlds_anom)-(1000*12):]
        rlus_eq = rlus_anom[len(rlus_anom)-(1000*12):]*-1
        hfnet_eq = hfnet_anom[len(hfnet_anom)-(1000*12):]
    else: 
        if exp.startswith("abrupt"):
            rlut_eq = rlut_anom[(150*12):]
            swcf_eq = swcf_anom[(150*12):]
            nino34_eq = nino34_anom[(150*12):]
            
            rsds_eq = rsds_anom[(150*12):]
            rsus_eq = rsus_anom[(150*12):]*-1
            rlds_eq = rlds_anom[(150*12):]
            rlus_eq = rlus_anom[(150*12):]*-1
            hfnet_eq = hfnet_anom[(150*12):]
        elif exp.startswith("1pct"):
            rlut_eq = rlut_anom[(140*12):]
            swcf_eq = swcf_anom[(140*12):]
            nino34_eq = nino34_anom[(140*12):]
            
            rsds_eq = rsds_anom[(140*12):]
            rsus_eq = rsus_anom[(140*12):]*-1
            rlds_eq = rlds_anom[(140*12):]
            rlus_eq = rlus_anom[(140*12):]*-1
            hfnet_eq = hfnet_anom[(140*12):]
        else:
            rlut_eq = rlut_anom
            swcf_eq = swcf_anom
            nino34_eq = nino34_anom
            
            rsds_eq = rsds_anom
            rsus_eq = rsus_anom*-1
            rlds_eq = rlds_anom
            rlus_eq = rlus_anom*-1
            hfnet_eq = hfnet_anom
            
    swnet = rsds_eq + rsus_eq.values # adding because upwelling already negative
    lwnet = rlds_eq + rlus_eq.values # adding because upwelling already negative
    netsurf = swnet + lwnet
    netflux = swnet + lwnet + (hfnet_eq.values*-1) # heat flux positive is out
    
    ## rolling regressions
    
    clouds_enso_exp = pd.DataFrame(np.transpose([rlut_eq.values,swcf_eq.values,rsds_eq.values,
                                                 rsus_eq.values,rlds_eq.values,rlus_eq.values,
                                                 swnet.values,lwnet.values,netsurf.values,netflux.values,nino34_eq.values]),
                                     columns=["rlut","swcf","rsds","rsus","rlds","rlus",
                                              "swnet","lwnet","netsurf","netflux","Nino34"])
    
    
    rlut_model = RollingOLS.from_formula("rlut ~ Nino34",data=clouds_enso_exp,window=100*12).fit()
    swcf_model = RollingOLS.from_formula("swcf ~ Nino34",data=clouds_enso_exp,window=100*12).fit()
    rsds_model = RollingOLS.from_formula("rsds ~ Nino34",data=clouds_enso_exp,window=100*12).fit()
    rsus_model = RollingOLS.from_formula("rsus ~ Nino34",data=clouds_enso_exp,window=100*12).fit()
    rlds_model = RollingOLS.from_formula("rlds ~ Nino34",data=clouds_enso_exp,window=100*12).fit()
    rlus_model = RollingOLS.from_formula("rlus ~ Nino34",data=clouds_enso_exp,window=100*12).fit()
    swnet_model = RollingOLS.from_formula("swnet ~ Nino34",data=clouds_enso_exp,window=100*12).fit()
    lwnet_model = RollingOLS.from_formula("lwnet ~ Nino34",data=clouds_enso_exp,window=100*12).fit()
    netsurf_model = RollingOLS.from_formula("netsurf ~ Nino34",data=clouds_enso_exp,window=100*12).fit()
    netflux_model = RollingOLS.from_formula("netflux ~ Nino34",data=clouds_enso_exp,window=100*12).fit()
    
    ## get coefficients
    
    beta_rlut_exp = rlut_model.params["Nino34"].values[~np.isnan(rlut_model.params["Nino34"].values)]
    beta_swcf_exp = swcf_model.params["Nino34"].values[~np.isnan(swcf_model.params["Nino34"].values)]
    beta_rsds_exp = rsds_model.params["Nino34"].values[~np.isnan(rsds_model.params["Nino34"].values)]
    beta_rsus_exp = rsus_model.params["Nino34"].values[~np.isnan(rsus_model.params["Nino34"].values)]
    beta_rlds_exp = rlds_model.params["Nino34"].values[~np.isnan(rlds_model.params["Nino34"].values)]
    beta_rlus_exp = rlus_model.params["Nino34"].values[~np.isnan(rlus_model.params["Nino34"].values)]
    beta_swnet_exp = swnet_model.params["Nino34"].values[~np.isnan(swnet_model.params["Nino34"].values)]
    beta_lwnet_exp = lwnet_model.params["Nino34"].values[~np.isnan(lwnet_model.params["Nino34"].values)]
    beta_netsurf_exp = netsurf_model.params["Nino34"].values[~np.isnan(netsurf_model.params["Nino34"].values)]
    beta_netflux_exp = netflux_model.params["Nino34"].values[~np.isnan(netflux_model.params["Nino34"].values)]
    
    beta_rlut_all.append(beta_rlut_exp)
    beta_swcf_all.append(beta_swcf_exp)
    beta_rsds_all.append(beta_rsds_exp)
    beta_rsus_all.append(beta_rsus_exp)
    beta_rlds_all.append(beta_rlds_exp)
    beta_rlus_all.append(beta_rlus_exp)
    beta_swnet_all.append(beta_swnet_exp)
    beta_lwnet_all.append(beta_lwnet_exp)
    beta_netsurf_all.append(beta_netsurf_exp)
    beta_netflux_all.append(beta_netflux_exp)
```

```python
beta_rsds_change = []
beta_rsds_exps = []
beta_rsds_ctrls = []
for i in np.arange(0,len(modelnames_file),1): 

    model, exp = modelnames_file[i].split("_")
    print(modelnames_file[i])
    
    rsds_control_anom = xr.DataArray(xr.open_dataset(loc_clouds+"rsds_nino34_"+model+"_control_anom_detrend.nc").data_vars["rsds"])
    nino34_control_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_control_anom_detrend.nc").data_vars["nino34"])

    rsds_exp_anom = xr.DataArray(xr.open_dataset(loc_clouds+"rsds_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["rsds"])
    nino34_exp_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino34"])
        
        
    if model == "GISSE2R":
        rsds_ctrl = rsds_control_anom[len(rsds_control_anom)-(1000*12):]
        nino34_ctrl = nino34_control_anom[len(nino34_control_anom)-(1000*12):]
        rsds_eq = rsds_exp_anom[len(rsds_exp_anom)-(1000*12):]
        nino34_eq = nino34_exp_anom[len(nino34_exp_anom)-(1000*12):]
    else:
        if exp.startswith("abrupt"):
            rsds_eq = rsds_exp_anom[(150*12):]
            nino34_eq = nino34_exp_anom[(150*12):]
            rsds_ctrl = rsds_control_anom
            nino34_ctrl = nino34_control_anom
        elif exp.startswith("1pct"):
            rsds_eq = rsds_exp_anom[(140*12):]
            nino34_eq = nino34_exp_anom[(140*12):]
            rsds_ctrl = rsds_control_anom
            nino34_ctrl = nino34_control_anom
        
        
    ## rolling regressions
    
    heatflux_enso_exp = pd.DataFrame(np.transpose([rsds_eq.values,nino34_eq.values]),
                                     columns=["rsds","Nino34"])
    heatflux_enso_control = pd.DataFrame(np.transpose([rsds_ctrl.values,nino34_ctrl.values]),
                                     columns=["rsds","Nino34"])
    
    rsds_model_exp = RollingOLS.from_formula("rsds ~ Nino34",data=heatflux_enso_exp,window=100*12).fit()
    rsds_model_ctrl = RollingOLS.from_formula("rsds ~ Nino34",data=heatflux_enso_control,window=100*12).fit()
    
    
    ## get coefficients
    
    beta_rsds_exp = rsds_model_exp.params["Nino34"].values[~np.isnan(rsds_model_exp.params["Nino34"].values)]
    beta_rsds_ctrl = rsds_model_ctrl.params["Nino34"].values[~np.isnan(rsds_model_ctrl.params["Nino34"].values)]
    
    beta_rsds_change_model = beta_rsds_exp - np.mean(beta_rsds_ctrl)
    beta_rsds_exps.append(beta_rsds_exp)
    beta_rsds_ctrls.append(beta_rsds_ctrl)
    beta_rsds_change.append(beta_rsds_change_model)
```
