Function for hiding right and top axis

```python
def hide_right_and_top(axis):
    
    # This function hides the right and top axes
    # of a given axis object
    # For purely aesthetic purposes
    
    # Hide the right and top spines
    axis.spines['right'].set_visible(False)
    axis.spines['top'].set_visible(False)

    # Only show ticks on the left and bottom spines
    axis.yaxis.set_ticks_position('left')
    axis.xaxis.set_ticks_position('bottom')
```

#### Analysis

Create figure object, loop through models, get indices and plot

```python
rm_yr = 100
rm_length = rm_yr*12
```

```python
fig = plt.figure(figsize=(36,26))

rcParams["font.family"] = "sans-serif"
rcParams["font.sans-serif"] = ["Helvetica Neue"]
rcParams["font.size"] = 23
rcParams["axes.linewidth"] = 2
#rcParams["axes.linewidth"] = 1.5
rcParams['xtick.major.size'] = 8
rcParams['xtick.major.width'] = 2
rcParams['ytick.major.size'] = 8
rcParams['ytick.major.width'] = 2

gs1 = gridspec.GridSpec(4,6)
gs1.update(left=0.1,right=0.96,top=0.93,bottom=0.1,wspace=0.47,hspace=0.42)

for i in np.arange(0,len(modelnames_file),1):
    
    if modelnames_file[i] == "MPIESM11_abrupt4x":
        rm_length = rm_yr
    else:
        rm_length = rm_yr*12
        
    model, exp = modelnames_file[i].split("_")
    #print(modelnames_file[i])
    
    if ((model == "CCSM3") | (model == "CESM104")):
        ax = plt.subplot(gs1[0,i])
    elif ((model == "GFDLCM3") | (model == "GFDLESM2M") | (model == "GISSE2R") | (model == "CNRMCM61")):
        ax = plt.subplot(gs1[1,i-6])
    elif ((model == "HadCM3L") | (model == "IPSLCM5A")):
        ax = plt.subplot(gs1[2,i-11])
    else:
        ax = plt.subplot(gs1[3,i-16])
    hide_right_and_top(ax)
    
    nino34_ctrl_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_control_anom_detrend2.nc").data_vars["nino34"])
    nino34_forced_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino34"])
    
    std_f1 = pd.Series(nino34_forced_anom).rolling(rm_length,center=False).std().values
    std_c1 = pd.Series(nino34_ctrl_anom).rolling(rm_length,center=False).std().values
    
    std_c = std_c1[~np.isnan(std_c1)]
    std_f = std_f1[~np.isnan(std_f1)]  
    
    if runtype[i] == "abrupt":
        amp_eq = std_f[150*12:]
        amp_trns = std_f[(50*12):(150*12)]
        amp_fast = std_f[0:(50*12)]
    else:
        amp_eq = std_f[140*12:]
        amp_trns = std_f[0:(140*12)]
        amp_fast = std_f[0:1]

    [d, p_t] = stats.ks_2samp(amp_trns,std_c)
    [d, p_e] = stats.ks_2samp(amp_eq,std_c)
    
    # Plot
    #sns.distplot(list(cmip5_occ),color="black",hist=False,kde=True,kde_kws={"linewidth": 3,"alpha":1},label="CMIP5 (18)")
    if modelnames_file[i] == "GISSE2R_abrupt4x":
    #if i == len(ex1)-1:
        d1 = sns.distplot(list(std_c),color="black",hist=False,kde=True,kde_kws={"linewidth":3,"alpha":1,"linestyle":"--"},label="Control")
        #p2 = plt.axvline(x=np.mean(amp_fast),color="gray",linewidth=3,linestyle="--",label="Fast (Pre-Transient)")
        p1 = plt.axvline(x=np.mean(amp_trns),color="red",linewidth=3,linestyle="-",label="Transient")
        d2 = sns.distplot(list(amp_eq),color="black",hist=False,kde=True,kde_kws={"linewidth":3,"alpha":1},label="Equilibrium")
        leg = plt.legend(frameon=False,loc="upper right",fontsize=35,bbox_to_anchor=(2.5, 0.8))
        for line in leg.get_lines():
            line.set_linewidth(5)
    else:
        d1 = sns.distplot(list(std_c),color="black",hist=False,kde=True,kde_kws={"linewidth":3,"alpha":1,"linestyle":"--"})
        d2 = sns.distplot(list(amp_eq),color="black",hist=False,kde=True,kde_kws={"linewidth":3,"alpha":1})
        #if runtype[i] == "abrupt":
            #p2 = plt.axvline(x=np.mean(amp_fast),color="gray",linewidth=3,linestyle="--")
        p1 = plt.axvline(x=np.mean(amp_trns),color="red",linewidth=3,linestyle="-")
        
    plt.xlim([-0.1,1.5])
    plt.xticks(ticks=[0,0.5,1,1.5])
    plt.ylabel("Density",labelpad=10)
    plt.yticks([])
    plt.xlabel("ENSO Amplitude ("+r"$\degree$C)")
    plt.title(modelnames_fig[i],fontsize=25)
    
    
plt.savefig("../Figures/FigS6_Amplitude_Density.pdf")
plt.show()
```
