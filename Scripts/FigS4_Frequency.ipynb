Function for setting boxplot color

```python
def set_boxplot_col(b,n,c,c_double,lw,median):
    
    # This function sets the colors of a given boxplot b
    # It auto-sets the colors of the boxes, whiskers, caps, and fliers to c[m]
    # and it sets the colors of the medians if median=True
    # n is the number of instances of the boxplot and lw is the desired
    # linewidth
    # c should be a vector of colors
    
    for m in range(0,n*2,1):
        if m < n:
            b["boxes"][m].set(color=c[m],linewidth=lw)
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
            #b["fliers"][m].set(markerfacecolor=c[m],markeredgecolor=c[m])
            if median==True:
                b["medians"][m].set(color=c[m],linewidth=lw)
        else:
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
```

Frequency calculation

```python
def frequency(x,t):
    
    # x = data
    # t = threshold
    n = len(x)
    
    xbool = x >= t
    
    count = 0
    for nn in np.arange(0,n-1,1):
        if ((xbool[nn] == 0) & (xbool[nn+1] == 1)):
            count = count + 1
            
    return(count)
```

Frequency calculation for an already-boolean array

```python
def frequency_bool(x):
    
    # x = data

    n = len(x)
        
    count = 0
    for nn in np.arange(0,n-1,1):
        if ((x[nn] == 0) & (x[nn+1] == 1)):
            count = count + 1
            
    return(count)
```

```python
ordr = 2

transient_change_ctrl = np.zeros(len(modelnames_fig))
equilibrium_change_ctrl = []
pvals_ctrl = np.zeros(len(modelnames_fig))

transient_change_forced = np.zeros(len(modelnames_fig))
equilibrium_change_forced = []
pvals_forced = np.zeros(len(modelnames_fig))

for j in np.arange(0,2,1):
    
    if j == 0:
        print("Independent control/forced thresholds")
    else:
        print("Single control thresholds")
        
    for i in np.arange(0,len(modelnames_file),1):
    
        model, exp = modelnames_file[i].split("_")
        print(modelnames_file[i])

        nino34_ctrl_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_control_anom_detrend.nc").data_vars["nino34"])
        nino34_forced_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino34"])
        
        # Smooth out the anomalies
        
        forced_anom1 = pd.Series(nino34_forced_anom).rolling(rm_length,center=False).mean().values
        ctrl_anom1 = pd.Series(nino34_ctrl_anom).rolling(rm_length,center=False).mean().values
    
        forced_anom = forced_anom1[~np.isnan(forced_anom1)]
        ctrl_anom = ctrl_anom1[~np.isnan(ctrl_anom1)]
        
        print("Calculating frequency...")
        
        # Set thresholds
        
        if j == 0: # separate control and forced thresholds
            
            ctrl_en_75 = np.percentile(ctrl_anom,75)
            ctrl_en_90 = np.percentile(ctrl_anom,90)
            ctrl_en = np.mean(ctrl_anom[(ctrl_anom >= ctrl_en_75) & (ctrl_anom <= ctrl_en_90)])
            
            forced_en_75 = np.percentile(forced_anom,75)
            forced_en_90 = np.percentile(forced_anom,90)
            forced_en = np.mean(forced_anom[(forced_anom >= forced_en_75) & (forced_anom <= forced_en_90)])
            
        else:
            
            ctrl_en_75 = np.percentile(ctrl_anom,75)
            ctrl_en_90 = np.percentile(ctrl_anom,90)
            ctrl_en = np.mean(ctrl_anom[(ctrl_anom >= ctrl_en_75) & (ctrl_anom <= ctrl_en_90)])
            
            forced_en_75 = np.percentile(forced_anom,75)
            forced_en_90 = np.percentile(forced_anom,90)
            #forced_en = np.mean(forced_anom[(forced_anom >= forced_en_75) & (forced_anom <= forced_en_90)])
            forced_en = ctrl_en
            
        # Boolean
        en_c = ctrl_anom >= ctrl_en
        en_f = forced_anom >= forced_en
        
        # Control frequency
        # using the "frequency" function we defined earlier plus "rolling_apply"
        freq_c1 = pd.Series(en_c).rolling(rm_length_freq).apply(frequency_bool,raw=True).values
        freq_c = freq_c1[~np.isnan(freq_c1)]
        
        
        
        # high-co2 frequency
        if runtype[i] == "abrupt":   
            en_eq = en_f[(150*12):]
            en_t = en_f[(50*12):(150*12)]
        else:       
            en_eq = en_f[(140*12):]
            en_t = en_f[0:(140*12)]
        
        freq_t = (frequency_bool(en_t))/(len(en_t)/1200)
        freq_e1 = pd.Series(en_eq).rolling(rm_length_freq).apply(frequency_bool,raw=True).values
        freq_e = freq_e1[~np.isnan(freq_e1)]
        
        # Calculate change in frequency
        freq_change_t = ((freq_t - np.mean(freq_c))/np.mean(freq_c))*100
        freq_change_e = ((freq_e - np.mean(freq_c))/np.mean(freq_c))*100
        
        #print("control frequency mean and SD:")
        #print(np.mean(freq_c))
        #print(np.std(freq_c))
        
        #print("forced frequency mean and SD:")
        #print(np.mean(freq_e))
        #print(np.std(freq_e))
        
        # Add to variables to save
        if j == 0:
            transient_change_forced[i] = freq_change_t
            equilibrium_change_forced.append(freq_change_e)
        else:
            transient_change_ctrl[i] = freq_change_t
            equilibrium_change_ctrl.append(freq_change_e)
```
