Function to set boxplot color

```python
def set_boxplot_col(b,n,c,c_double,lw,median):
    
    # This function sets the colors of a given boxplot b
    # It auto-sets the colors of the boxes, whiskers, caps, and fliers to c[m]
    # and it sets the colors of the medians if median=True
    # n is the number of instances of the boxplot and lw is the desired
    # linewidth
    # c should be a vector of colors
    
    for m in range(0,n*2,1):
        if m < n:
            b["boxes"][m].set(color=c[m],linewidth=lw)
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
            #b["fliers"][m].set(markerfacecolor=c[m],markeredgecolor=c[m])
            if median==True:
                b["medians"][m].set(color=c[m],linewidth=lw)
        else:
            b["whiskers"][m].set(color=c_double[m],linewidth=lw)
            b["caps"][m].set(color=c_double[m],linewidth=lw)
```

#### Analysis

Read data and calculate feedbacks

```python
beta_netflux_change = []
beta_netflux_exps = []
beta_netflux_ctrls = []
beta_netflux_change_pct_em = np.zeros(len(modelnames_file))
beta_netflux_change_em = np.zeros(len(modelnames_file))

for i in np.arange(0,len(modelnames_file),1): 

    model, exp = modelnames_file[i].split("_")
    print(modelnames_file[i])
    
    rsds_control_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rsds_nino34_"+model+"_control_anom_detrend.nc").data_vars["rsds"])
    rsus_control_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rsus_nino34_"+model+"_control_anom_detrend.nc").data_vars["rsus"])
    rlds_control_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rlds_nino34_"+model+"_control_anom_detrend.nc").data_vars["rlds"])
    rlus_control_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rlus_nino34_"+model+"_control_anom_detrend.nc").data_vars["rlus"])
    hfnet_control_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfnet_nino34_"+model+"_control_anom_detrend.nc").data_vars["hfnet"])
    nino34_control_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_control_anom_detrend.nc").data_vars["nino34"])

    rsds_exp_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rsds_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["rsds"])
    rsus_exp_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rsus_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["rsus"])
    rlds_exp_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rlds_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["rlds"])
    rlus_exp_anom = xr.DataArray(xr.open_dataset(loc_radflux+"rlus_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["rlus"])
    hfnet_exp_anom = xr.DataArray(xr.open_dataset(loc_hf+"hfnet_nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["hfnet"])
    nino34_exp_anom = xr.DataArray(xr.open_dataset(loc_nino34+"nino34_"+model+"_"+exp+"_anom_detrend2.nc").data_vars["nino34"])
        
        
    if model == "GISSE2R":
        rsds_ctrl = rsds_control_anom[len(rsds_control_anom)-(1000*12):]
        rsus_ctrl = rsus_control_anom[len(rsus_control_anom)-(1000*12):]
        rlds_ctrl = rlds_control_anom[len(rlds_control_anom)-(1000*12):]
        rlus_ctrl = rlus_control_anom[len(rlus_control_anom)-(1000*12):]
        hfnet_ctrl = hfnet_control_anom[len(hfnet_control_anom)-(1000*12):]
        nino34_ctrl = nino34_control_anom[len(nino34_control_anom)-(1000*12):]
        
        rsds_eq = rsds_exp_anom[len(rsds_exp_anom)-(1000*12):]
        rsus_eq = rsus_exp_anom[len(rsus_exp_anom)-(1000*12):]
        rlds_eq = rlds_exp_anom[len(rlds_exp_anom)-(1000*12):]
        rlus_eq = rlus_exp_anom[len(rlus_exp_anom)-(1000*12):]
        hfnet_eq = hfnet_exp_anom[len(hfnet_exp_anom)-(1000*12):]
        nino34_eq = nino34_exp_anom[len(nino34_exp_anom)-(1000*12):]
    else:
        if exp.startswith("abrupt"):
            rsds_eq = rsds_exp_anom[(150*12):]
            rsus_eq = rsus_exp_anom[(150*12):]
            rlds_eq = rlds_exp_anom[(150*12):]
            rlus_eq = rlus_exp_anom[(150*12):]
            hfnet_eq = hfnet_exp_anom[(150*12):]
            nino34_eq = nino34_exp_anom[(150*12):]
        elif exp.startswith("1pct"):
            rsds_eq = rsds_exp_anom[(140*12):]
            rsus_eq = rsus_exp_anom[(140*12):]
            rlds_eq = rlds_exp_anom[(140*12):]
            rlus_eq = rlus_exp_anom[(140*12):]
            hfnet_eq = hfnet_exp_anom[(140*12):]
            nino34_eq = nino34_exp_anom[(140*12):]
            
        rsds_ctrl = rsds_control_anom
        rsus_ctrl = rsus_control_anom
        rlds_ctrl = rlds_control_anom
        rlus_ctrl = rlus_control_anom
        hfnet_ctrl = hfnet_control_anom
        nino34_ctrl = nino34_control_anom
    
    #swnet = rsds_eq + rsus_eq.values # adding because upwelling already negative
    #lwnet = rlds_eq + rlus_eq.values # adding because upwelling already negative
    #netsurf = swnet + lwnet
    #netflux = swnet + lwnet + (hfnet_eq.values*-1) # heat flux positive is out
    
    # heat flux positive is out
    # upwelling is negative relative to downwelling
    netflux_eq = rsds_eq + rlds_eq.values - rsus_eq.values - rlus_eq.values - hfnet_eq.values
    netflux_ctrl = rsds_ctrl + rlds_ctrl.values - rsus_ctrl.values - rlus_ctrl.values - hfnet_ctrl.values
    
    
    ## rolling regressions
    
    heatflux_enso_exp = pd.DataFrame(np.transpose([netflux_eq.values,nino34_eq.values]),
                                     columns=["netflux","Nino34"])
    heatflux_enso_control = pd.DataFrame(np.transpose([netflux_ctrl.values,nino34_ctrl.values]),
                                     columns=["netflux","Nino34"])
    
    netflux_model_exp = RollingOLS.from_formula("netflux ~ Nino34",data=heatflux_enso_exp,window=100*12).fit()
    netflux_model_ctrl = RollingOLS.from_formula("netflux ~ Nino34",data=heatflux_enso_control,window=100*12).fit()
    
    
    ## get coefficients
    
    beta_netflux_exp = netflux_model_exp.params["Nino34"].values[~np.isnan(netflux_model_exp.params["Nino34"].values)]
    beta_netflux_ctrl = netflux_model_ctrl.params["Nino34"].values[~np.isnan(netflux_model_ctrl.params["Nino34"].values)]
    
    beta_netflux_change_model = beta_netflux_exp - np.mean(beta_netflux_ctrl)
    beta_netflux_exps.append(beta_netflux_exp)
    beta_netflux_ctrls.append(beta_netflux_ctrl)
    beta_netflux_change.append(beta_netflux_change_model)
    
    beta_netflux_change_pct_model = 100*(beta_netflux_exp - np.mean(beta_netflux_ctrl))/np.mean(beta_netflux_ctrl)
    beta_netflux_change_pct_em[i] = np.mean(beta_netflux_change_pct_model)
    
    beta_netflux_change_em[i] = np.mean(beta_netflux_change_model)
```

Plot

```python
fig = plt.figure(figsize=(24,10))

rcParams["font.family"] = "sans-serif"
rcParams["font.sans-serif"] = ["Helvetica Neue"]
rcParams["font.size"] = 30.0
rcParams["axes.linewidth"] = 1.5
rcParams['xtick.major.size'] = 8
rcParams['xtick.major.width'] = 1.5
rcParams['ytick.major.size'] = 8
rcParams['ytick.major.width'] = 1.5

gs1 = gridspec.GridSpec(1,1)
gs1.update(left=0.15,right=0.92,top=0.95,bottom=0.4,wspace=0.2,hspace=0.28)

beta = beta_netflux_change

ax = plt.subplot(gs1[0,0])

b1 = plt.boxplot(beta,whis=[5,95],sym="",patch_artist=True)
set_boxplot_col(b1,len(modelnames_file),np.array(colors)/255.,np.array(colors_double)/255.,3,True)
for patch in b1["boxes"]:
    patch.set(facecolor="white")

plt.plot([1,2,3],[np.median(beta[0]),np.median(beta[1]),np.median(beta[2])],linewidth=2.5,color=np.array(colors[0])/255.)
plt.plot([8,9,10,11],[np.median(beta[7]),np.median(beta[8]),np.median(beta[9]),np.median(beta[10])],linewidth=2.5,color=np.array(colors[8])/255.)
plt.plot([15,16,17],[np.median(beta[14]),np.median(beta[15]),np.median(beta[16])],linewidth=2.5,color=np.array(colors[14])/255.)

plt.axhline(y=0,linewidth=1.5,color="k",linestyle="-")

plt.xticks(np.arange(1,len(modelnames_fig)+1,1),modelnames_fig,
           rotation=45,horizontalalignment="right",rotation_mode="anchor")

plt.xlim([0,len(modelnames_file)+1])
#plt.ylim([-8.5,5.5])
plt.ylabel("Change in net thermodynamic"+"\n"+" feedback (Wm$^{-2}$K$^{-1}$)")

plt.savefig("../Figures/FigS15_NetFlux_Feedback.pdf")
plt.show()
```
